```python
import numpy as np
import matplotlib.pyplot as plt

import pandas as pd

plt.rcParams["figure.dpi"] = 130

dataFolder = '/eno/jdfeathe/DATA/SpaceJam/force_data/'
```

```python
file = '2022-03-16_Medium.CSV'

df = pd.read_csv(dataFolder + file)

for i in range(4):
    plt.plot( df.iloc[:,i+1])
    
plt.show()
```

```python
b = [100, 200]

forceArr = np.abs(df.iloc[b[0]:b[1],1] - df.iloc[b[0],1])
tArr = df.iloc[b[0]:b[1],0]

plt.plot(tArr, forceArr)
plt.show()
```
